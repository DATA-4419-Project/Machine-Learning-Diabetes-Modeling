```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/Users/cartersilos/Downloads/DATASET_clean.csv")
```

```python
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Assume df is your DataFrame with numeric columns only
corr_matrix = df.corr()

# Mask self-correlations and duplicate pairs
mask = np.tril(np.ones(corr_matrix.shape), k=0).astype(bool)
corr_matrix_masked = corr_matrix.mask(mask)

# Flatten to Series and drop NaNs from the mask
corr_pairs = corr_matrix_masked.unstack().dropna()

# Most positively correlated pair
most_pos_pair = corr_pairs.idxmax()
most_pos_value = corr_pairs.max()

# Most negatively correlated pair
most_neg_pair = corr_pairs.idxmin()
most_neg_value = corr_pairs.min()

# Least correlated pair (closest to 0)
least_corr_pair = corr_pairs.abs().idxmin()
least_corr_value = corr_pairs[least_corr_pair]

# Output
print(f" Most positively correlated(closest to +1): {most_pos_pair} = {most_pos_value:.4f}")
print(f" Most negatively correlated (closest to -1): {most_neg_pair} = {most_neg_value:.4f}")
print(f" Least correlated (closest to 0): {least_corr_pair} = {least_corr_value:.4f}")
```
