```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
A1C_DATASET_E = pd.read_csv("A1C_DATASET_E.csv")
```

```python
A1C_DATASET_E
```

```python
a1c_data = A1C_DATASET_E.dropna()
```

```python
a1c_data.shape
a1c_data.columns
```

```python
a1c_data['a1c_test'] = np.where(a1c_data['LBXGH'] > 6.4, 1, 0)
```

```python
a1c_data[['SEQN','a1c_test', 'LBXGH']]
```

```python
a1c_data
```

```python
X = a1c_data.drop(columns = ['SEQN', 'LBXGH', 'a1c_test', 'Unnamed: 0', 'DIQ010' ])
```

```python
X['DPQ010'] = X['DPQ010'].round().astype(int)
X['DPQ040'] = X['DPQ040'].round().astype(int)
X['DPQ050'] = X['DPQ050'].round().astype(int)
X['DPQ060'] = X['DPQ060'].round().astype(int)
X['DMDBORN'] = X['DMDBORN'].round().astype(int)
X['DMDEDUC'] = X['DMDEDUC'].round().astype(int)
X['DMDMARTX'] = X['DMDMARTX'].round().astype(int)
X['ALQ111'] = X['ALQ111'].round()
X['PAD790Q'] = a1c_data['PAD790Q'].round() ## NOT A CATEGORY
X['PAD680'] = X['PAD680'].round()          ## NOT A CATEGORY
X['SMQ020'] = X['SMQ020'].round().astype(int)
#X['SEQN'] = X['SEQN'].round().astype(int)
```

```python
X_factors = ['ALQ111', 'SMQ020', 'DPQ010', 'DPQ040', 'DPQ050', 'DPQ060', 'DMDBORN', 'DMDEDUC', 'DMDMARTX']
X[X_factors] = X[X_factors].astype('category')
X
```

```python
#X_indexed = X.set_index('SEQN')
#X_indexed.to_csv('X_indexed.csv')
```

```python
#X_indexed
```

```python
y = a1c_data['a1c_test']
y_indexed = a1c_data[['SEQN', 'a1c_test']]
y_indexed['SEQN'] = y_indexed['SEQN'].round().astype(int)
y_indexed = y_indexed.set_index('SEQN')
y_indexed.to_csv('y_indexed.csv')
```

```python
y
```

```python
X.to_csv('X.csv', index=False)
y.to_csv('y.csv', index=False)
```

```python
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

# l1 LASSO, C=100, solver=SAGA
lr = LogisticRegression(penalty='l1', C=1, random_state=0, solver='saga', max_iter = 10000)
lr.fit(X_train, y_train)
```

```python
from sklearn import metrics

# generate evaluation metrics
print("Train - Accuracy :", metrics.accuracy_score(y_train, lr.predict(X_train)))
print("Train - Confusion matrix :",metrics.confusion_matrix(y_train, lr.predict(X_train)))
print("Train - classification report :", metrics.classification_report(y_train, lr.predict(X_train)))

print("Test - Accuracy :", metrics.accuracy_score(y_test, lr.predict(X_test)))
print("Test - Confusion matrix :",metrics.confusion_matrix(y_test, lr.predict(X_test)))
print("Test - classification report :", metrics.classification_report(y_test, lr.predict(X_test)))
```
